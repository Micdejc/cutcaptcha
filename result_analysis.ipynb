## Overall stats of test execution

```python
# Import libraries as required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 10)

# Load in the data set
df = pd.read_csv('./full_results.csv')
df
```

```python
count=df["Correct?"].value_counts()
print(count)
labels= ['True', 'False', 'NaN']
colors=['blue', 'orange', 'green']
df["Correct?"].value_counts(normalize=True).plot.pie(labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')
```

```python
# Prepare data
x = df['Test']
y = df['Time(ms)'] / 1000
y_mean = np.mean(y)

# IEEE single-column size
plt.figure(figsize=(4.0, 2.5))

# Plot data (colorblind-friendly blue)
plt.plot(x, y,
         linestyle='-',
         marker='',
         markersize=1,
         linewidth=0.8,
         color='#0072B2',
         label='Execution Time')

# Plot mean (distinct color + linestyle)
plt.axhline(y=y_mean,
            linestyle='--',
            linewidth=1,
            color='#D55E00',
            label=f'Mean = {y_mean:.2f} s')

# Labels
plt.xlabel("Test Number", fontsize=8)
plt.ylabel("Time (s)", fontsize=8)

# Title (optional)
#plt.title("Execution Time Across Tests", fontsize=9)

# Ticks
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

# Grid
plt.grid(True, linestyle=':', linewidth=0.5)

# Legend
plt.legend(fontsize=7, frameon=False)

# Layout
plt.tight_layout()

# Save
#plt.savefig("execution_time_plot.pdf")
plt.savefig("execution_time_plot.png", dpi=600)

plt.show()
plt.close()
```

## Test execution with non-similarity in shapes 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)

# We try to find how many challenges were fragments with distinct shapes by extracting the subset of challenges...
# using fragments with similar shapes from the whole test set
all_df = pd.read_csv("./full_results.csv")
subset_df = pd.read_csv("./results_with_similar_fragments.csv")

# Remove rows where 'Test(id)' exists in subset
result_df = all_df[~all_df['Test'].isin(subset_df['Test'])]

result_df.to_csv("results_with_dissimilar_fragments.csv", index=False)

result_df

count=result_df["Correct?"].value_counts()
print(count)
labels= ['True', 'False', 'NaN']
colors=['blue', 'orange', 'green']
result_df["Correct?"].value_counts(normalize=True).plot.pie(labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')
```
