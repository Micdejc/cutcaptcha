```python
# Import libraries as required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)

# Load in the data set
df = pd.read_csv('./ResultsMscProject.csv')
df
```

```python
count=df["Correct?"].value_counts()
print(count)
labels= ['True', 'False', 'NaN']
colors=['blue', 'orange', 'green']
df["Correct?"].value_counts(normalize=True).plot.pie(labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')
```

```python
fig = plt.figure(figsize=(20,10))
# Calculate the simple average of the data
x=df['Test']
y=df['Time(ms)']/1000
y_mean = [np.mean(y)]*len(x)
ax = fig.add_subplot()
ax.plot(x, y, color='blue')
ax.plot(x, y_mean, color='orange')
ax.legend(['Data', 'Mean'])
ax.set_xlabel("Test number")
ax.set_ylabel("Solving time(s)")
ax.set_title("Tests line chart (mean time: %.2f seconds)"%np.mean(y))
```
